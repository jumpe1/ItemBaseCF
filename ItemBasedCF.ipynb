```python
import pandas as pd

# ratings_cols = ['user_id', 'movie_id', 'rating']
# df_ratings = pd.read_csv('/content/drive/My Drive/research/ml-1m/ratings.dat', sep='::', names=ratings_cols, usecols=range(3))
# movies_cols = ['movie_id', 'title']
# df_movies = pd.read_csv('/content/drive/My Drive/research/ml-1m/movies.dat', sep='::', names=movies_cols, usecols=range(2))

ratings_cols = ['user_id', 'movie_id', 'rating']
df_ratings = pd.read_csv('ml-1m/ratings.dat', sep='::', names=ratings_cols, usecols=range(3))
movies_cols = ['movie_id', 'title']
df_movies = pd.read_csv('ml-1m/movies.dat', sep='::', names=movies_cols, usecols=range(2))

ratings = pd.merge(df_ratings, df_movies)
ratings.head()
```

```python
userRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
userRatings
```

```python
corrMatrix = userRatings.corr()
corrMatrix
```

```python
corrMatrix = userRatings.corr(method='pearson', min_periods=150)
corrMatrix
```

```python
myRatings = userRatings.loc[0].dropna()
myRatings
```

```python
simCandidates = pd.Series()
for i in range(0, len(myRatings.index)):
#     print ("Adding sims for " + myRatings.index[i] + "...")

#         相関マトリックスから評点した映画と似たデータを取り出す　
    sims = corrMatrix[myRatings.index[i]].dropna()
#     その相関スコアと評価値をかけてあげることでおすすめ度とする
#     sims = sims.map(lambda x: x * myRatings[i])
    sims = sims * myRatings[i]
#         おすすめ映画に加える
    simCandidates = simCandidates.append(sims)
    
# print ("sorting...")
simCandidates.sort_values(inplace = True, ascending = False)
print (simCandidates.head(10))
```

```python
simCandidates = simCandidates.groupby(simCandidates.index).sum()
```

```python
simCandidates.sort_values(inplace = True, ascending = False)
simCandidates.head(10)
```

```python
# simCandidates.head(10)myRatings
filteredSims = simCandidates.drop(myRatings.index)
filteredSims.head(30)
```
